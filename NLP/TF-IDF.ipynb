```python
import pandas as pd
messages = pd.read_csv('spam.csv', 
                        encoding='latin-1', 
                        usecols=[0, 1], 
                        names=["label", "message"], 
                        header=0)
```

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
ps = WordNetLemmatizer()



corpus = []

for i in range (0, len(messages)):
    review = re.sub('[^a-zA-Z]',' ', str(messages['message'][i]))
    review = review.lower()
    review = review.split()
    review = [ps.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```
